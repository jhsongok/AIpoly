# DataFrame 병합하기

```python
import pandas as pd
import numpy as np
```

#### concat 함수를 사용하여 DataFrame 병합
 - 축을 따라 dataframe을 병합
   + axis=0 - 행단위 병합
   + axis=1 - 열단위 병합

#### column명이 같은 경우

```python
df1 = pd.DataFrame({'key1' : np.arange(5), 'value1' : np.random.randn(5)})
```

```python
df2 = pd.DataFrame({'key1' : np.arange(5), 'value1' : np.random.randn(5)})
```

```python
df2
```

```python
pd.concat([df1, df2])  # 행 레벨(axis=0)로 병합. 기존 index값은 그대로 유지
```

```python
pd.concat([df1, df2], ignore_index=True)  # 행 레벨(axis=0)로 병합. 기존 index값을 무시하고 새롭게 생성
```

```python
pd.concat([df1, df2], axis=1)  # 열 레벨(axis=1)로 병합
```

#### column 명이 다른 경우 

```python
df3 = pd.DataFrame({'key2' : np.arange(5), 'value2' : np.random.randn(5)})
```

```python
pd.concat([df1, df3]) # 행 레벨(axis=0)로 병합
```

```python
pd.concat([df1, df3], axis=1) # 열 레벨(axis=1)로 병합
```
