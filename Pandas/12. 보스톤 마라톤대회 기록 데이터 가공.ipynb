```python
import pandas as pd
```

#### 1. 원본 파일 읽어오기(DataFrame 생성)

```python
marathon_2015 = pd.read_csv("../marathon_data/marathon_results_2015.csv")
marathon_2016 = pd.read_csv("../marathon_data/marathon_results_2016.csv")
marathon_2017 = pd.read_csv("../marathon_data/marathon_results_2017.csv")
```

```python
marathon_2017.head()
```

#### 2. DataFrame에 column 추가

```python
marathon_2015['Year'] = '2015'
marathon_2016['Year'] = '2016'
marathon_2017['Year'] = '2017'
```

#### 3. DataFrame 병합하기

```python
marathon_2015_2017 = pd.concat([marathon_2015, marathon_2016, marathon_2017], ignore_index=True, sort=False)
```

```python
marathon_2015_2017.info()
```

#### 4. 불필요한 column 제거하기

```python
marathon_2015_2017 = marathon_2015_2017.drop(['Unnamed: 0','Bib', 'Citizen', 'Unnamed: 9', 'Proj Time', 'Unnamed: 8'], axis=1)
```

#### 5. column 값 변환하기

```python
# 시간 값(문자열)을 timedelta 타입으로 변환
marathon_2015_2017['5K'] = marathon_2015_2017['5K'].str.rjust(8, fillchar='0')
marathon_2015_2017['5K'] = pd.to_timedelta(marathon_2015_2017['5K'])
marathon_2015_2017['10K'] = marathon_2015_2017['10K'].str.rjust(8, fillchar='0')
marathon_2015_2017['10K'] = pd.to_timedelta(marathon_2015_2017['10K'])
marathon_2015_2017['15K'] = marathon_2015_2017['15K'].str.rjust(8, fillchar='0')
marathon_2015_2017['15K'] = pd.to_timedelta(marathon_2015_2017['15K'])
marathon_2015_2017['20K'] = marathon_2015_2017['20K'].str.rjust(8, fillchar='0')
marathon_2015_2017['20K'] = pd.to_timedelta(marathon_2015_2017['20K'])
marathon_2015_2017['Half'] = marathon_2015_2017['Half'].str.rjust(8, fillchar='0')
marathon_2015_2017['Half'] = pd.to_timedelta(marathon_2015_2017['Half'])
marathon_2015_2017['25K'] = marathon_2015_2017['25K'].str.rjust(8, fillchar='0')
marathon_2015_2017['25K'] = pd.to_timedelta(marathon_2015_2017['25K'])
marathon_2015_2017['30K'] = marathon_2015_2017['30K'].str.rjust(8, fillchar='0')
marathon_2015_2017['30K'] = pd.to_timedelta(marathon_2015_2017['30K'])
marathon_2015_2017['35K'] = marathon_2015_2017['35K'].str.rjust(8, fillchar='0')
marathon_2015_2017['35K'] = pd.to_timedelta(marathon_2015_2017['35K'])
marathon_2015_2017['40K'] = marathon_2015_2017['40K'].str.rjust(8, fillchar='0')
marathon_2015_2017['40K'] = pd.to_timedelta(marathon_2015_2017['40K'])
marathon_2015_2017['Pace'] = marathon_2015_2017['Pace'].str.rjust(8, fillchar='0')
marathon_2015_2017['Pace'] = pd.to_timedelta(marathon_2015_2017['Pace'])
marathon_2015_2017['Official Time'] = marathon_2015_2017['Official Time'].str.rjust(8, fillchar='0')
marathon_2015_2017['Official Time'] = pd.to_timedelta(marathon_2015_2017['Official Time'])
```

```python
marathon_2015_2017.info()
```

```python
# 시간(time)을 초(seconds)로 변경
marathon_2015_2017['5K'] = marathon_2015_2017['5K'].dt.total_seconds().astype('int64')
marathon_2015_2017['10K'] = marathon_2015_2017['10K'].dt.total_seconds().astype('int64')
marathon_2015_2017['15K'] = marathon_2015_2017['15K'].dt.total_seconds().astype('int64')
marathon_2015_2017['20K'] = marathon_2015_2017['20K'].dt.total_seconds().astype('int64')
marathon_2015_2017['Half'] = marathon_2015_2017['Half'].dt.total_seconds().astype('int64')
marathon_2015_2017['25K'] = marathon_2015_2017['25K'].dt.total_seconds().astype('int64')
marathon_2015_2017['30K'] = marathon_2015_2017['30K'].dt.total_seconds().astype('int64')
marathon_2015_2017['35K'] = marathon_2015_2017['35K'].dt.total_seconds().astype('int64')
marathon_2015_2017['40K'] = marathon_2015_2017['40K'].dt.total_seconds().astype('int64')
marathon_2015_2017['Pace'] = marathon_2015_2017['Pace'].dt.total_seconds().astype('int64')
marathon_2015_2017['Official Time'] = marathon_2015_2017['Official Time'].dt.total_seconds().astype('int64')
```

#### 6. 가공된 DataFrame을 csv파일로 저장하기

```python
marathon_2015_2017.to_csv("../marathon_data/marathon_2015_2017.csv", index = None, header=True)
```
