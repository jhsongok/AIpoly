# DataFrame 에서 원하는 데이터만 가져오기

#### 원하는 column만 선택해 가져오기
  - column 을 하나만 명시하는 경우
      - .(점)다음에 column 명시, 결과는 Series
      - [  ] 대괄호 안에 column 명시, 결과는 Series
  - column 리스트를 전달하는 경우
      + [  ] 대괄호 안에 **column 리스트**를 명시, 결과는 DataFrame
 

```python
import pandas as pd
```

```python
marathon_2017 = pd.read_csv("../marathon_data/marathon_results_2017.csv")
marathon_2017.head()
```

#### column을 하나만 명시하는 경우1) . (점) 다음에 column 명시

```python
marathon_2017.Name  # Series 형식으로 가져온다.
```

#### column을 하나만 명시하는 경우2) [   ] 안에 column 명시

```python
marathon_2017['Name']   # Series 형식으로 가져온다.
```

#### column 리스트를 전달하는 경우

```python
marathon_2017[['Name', 'Age', 'M/F', 'City', 'Country']]  # Column 리스트를 명시함, DataFrame 형식으로 가져온다.
```

```python
marathon_2017[['Country']] # Column 리스트를 명시함, DataFrame 형식으로 가져온다.
```

#### 원하는 row 만 선택해 가져오기
  - **슬라이싱**으로 데이터 가져오기     
  - **.loc**, **.iloc 함수**로 row를 선택해 가져오기

#### 슬라이싱으로 데이터 가져오기
  - Series의 경우 [  ] 를 사용에 row 선택이 가능하나, DataFrame의 경우는 [  ] 은 column 선택에만 사용함
  - 슬라이싱은 row 레벨로만 지원함

```python
marathon_2017[:10]  # index 0번째 부터 10을 포함하지 않은 row를 가져온다.
```

```python
marathon_2017[7:10]
```

#### .loc, .iloc 함수로 row를 선택해 가져오기
  + loc 함수 - index 자체를 사용함
  + iloc 함수 - 0 based index를 사용함
  + ,(콤마)를 사용하여 원하는 column도 동시 선택 가능

```python
marathon_2017.info()
```

```python
import numpy as np

marathon_2017.index = np.arange(100, 26510)  # 인덱스를 재정의 함
```

```python
marathon_2017.head()
```

```python
marathon_2017.loc[103]  # 하나만 명시하면 Series 형식으로 가져온다.
```

```python
marathon_2017.loc[[103, 307, 509, 997]]  # 리스트로 명시하면 DataFrame 형식으로 가져온다.
```

```python
marathon_2017.iloc[[0, 100, 200, 450]]  # 기존 인덱스와 상관없이 0 Based index로 가져오고 싶은 경우 iloc 를 사용한다.
```

#### 원하는 row와 column 동시 선택해 가져오기

```python
marathon_2017.loc[[103, 307, 509, 997], ['Name', 'Age', 'M/F', 'City', 'Country']] # 앞에는 row 를 명시하고 뒤에는 컬럼을 명시
```

```python
marathon_2017.iloc[[0, 100, 200, 450], [2, 3, 4, 5, 7]] # 앞에는 row 를 명시하고 뒤에는 컬럼을 명시
                                                        # iloc 에서는 row 는 0 based index를 명시해 주어야 하며
                                                        #  column 은  0 based column 을 명시해 주어야 함
```

#### boolean selection 으로 row 선택해 가져오기
  - 해당 조건에 맞는 row만 선택적으로 가져옴
  - 다중 조건의 경우 각각의 조건은 ()괄호로 묶어주고 &(and), |(or)를 사용하여 연결

```python
Country_ = marathon_2017['Country'] == 'KOR'
Age_ = (marathon_2017['Age'] >= 30) & (marathon_2017['Age'] < 40)

marathon_2017[Country_ & Age_]
```
