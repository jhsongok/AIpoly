# DataFrame 데이터 확인하기

```python
import pandas as pd
```

```python
marathon_2017 = pd.read_csv("../marathon_data/marathon_results_2017.csv")
marathon_2017.head()
```

#### index 속성
 - DataFrame의 행 인덱스 배열
 - 각 개별 데이터(row)를 특정할 수 있는 고유 값이 저장되어 있음

```python
marathon_2017.index
```

#### column 속성
 - DataFrame의 열 이름(column) 배열
 - 개별 속성/특성(feature)을 나타냄

```python
marathon_2017.columns
```

#### head, tail 함수
 - 데이터 전체가 아닌, 일부(처음부터, 혹은 마지막부터)를 간단히 보기 위한 함수

```python
marathon_2017.head(n=3) # n을 정의하지 않으면 기본적으로 5개만 보여줌
```

```python
marathon_2017.tail(n=4) # n을 정의하지 않으면 기본적으로 5개만 보여줌
```

#### sort_values 함수
  - 특정한 column의 값을 중심으로 ascending(오름차순)/descending(내림차순) 정렬
  - ascending(오름차순) : ascending=True 또는 생략(default)
  - descending(내림차순) : ascending=False

```python
marathon_2017.sort_values(by=['Age'])    # Age ascending 정렬
```

```python
marathon_2017.sort_values(by=['Age'], ascending=False)    # Age descending 정렬
```

#### shape 속성(row, column)

```python
marathon_2017.shape   # row, colume, 26410개의 row, 25개의 속성(colume)이 있음을 확인
```

#### describe 함수
 - 숫자형 데이터 타입 column들의 기본적인 통계치를 보여줌

```python
marathon_2017.describe()  # 숫자형 데이터 타입 column들에 대해 기본적인 통계치를 보여줌
```

 #### info 함수 
  - DataFrame 에 대한 기본적인 정보를 보여준다(row 개수, column 개수, 각 column들의 데이터 타입 등)

```python
marathon_2017.info()  # 데이터 타입과 개수를 확인해 볼 수 있음
```

#### column의 데이터 타입
  - 숫자형(Numerical Type) : 정수형(int), 실수형(float), 뒤의 숫자는 byte의 크기
  - 범주형(Categorical Type) : 문자열(object)
  - 타입 변경은 **astype 함수**를 사용하여 변경

```python
# 숫자형 데이터 타입을 범주형으로 변환
marathon_2017['Age'] = marathon_2017['Age'].astype(str)
marathon_2017.info()
```

```python
# 범주형 데이터 타입을 숫자형으로 변환
import numpy as np

marathon_2017['Age'] = marathon_2017['Age'].astype(np.int64)
marathon_2017.info()
```
