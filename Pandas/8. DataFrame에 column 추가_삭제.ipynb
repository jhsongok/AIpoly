# DataFrame 에 column 추가 및 삭제

```python
import pandas as pd
```

```python
marathon_2017 = pd.read_csv("../marathon_data/marathon_results_2017.csv")
marathon_2017.head()
```

#### column 추가하기
 - [  ] (대괄호) 안에 기존에 없는 새로운 column 이름을 부여해 column 을 추가할 수 있으며 맨 마지막에 추가 됨
 - 만약 [  ] (대괄호) 안에 명시한 column이 존재하는 column이면 해당 column이 업데이트 됨
 - insert 함수를 사용하면 원하는 위치에 column 추가

```python
marathon_2017['Age_double'] = marathon_2017['Age'] * 2  # 기존에 없는 새로운 이름을 부여하여 새 Coloum 생성, 맨 마지막에 생성됨
marathon_2017.head()
```

```python
marathon_2017['Age_tripple'] = marathon_2017['Age_double'] + marathon_2017['Age']
marathon_2017.head()
```

```python
# insert 함수를 사용하면 원하는 column 위치에 생성시킬 수 있다
marathon_2017.insert(3, 'Address', marathon_2017['Country']+ ' ' + marathon_2017['City'])
marathon_2017.head()
```

#### column 삭제하기
 - drop 함수 사용하여 삭제
   - axis=1 : 열 레벨
   - colunm 리스트를 사용하면 멀티플 삭제 가능 
   - inplace=True 를 설정하면 원본 DataFrame에서 삭제됨

```python
marathon_2017.drop('Age_tripple', axis=1)  # axis=1 이면 열 레벨
```

```python
marathon_2017.head()  # 위에서 Age_tripple column 을 지웠으나 다시 실행해 보니 지워지 않았음.
                      # drop 함수는 원본 DataFrame 을 변경하지 않고 새로운 DataFrame 복사본을 생성하여 
                      # Age_tripple column을 삭제 후 복사본의 결과를 보여줌
```

```python
marathon_2017.drop(['Address', 'Age_double', 'Age_tripple'], axis=1)
```

```python
marathon_2017.head()
```

```python
# inplace=True 를 설정하면 원본 DataFrame에서 삭제됨
marathon_2017.drop(['Address', 'Age_double', 'Age_tripple'], axis=1, inplace=True)
```
